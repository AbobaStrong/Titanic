```python
import numpy as np 
import pandas as pd

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

x_train = train_data.drop(['Survived'], axis = 1)
y_train = train_data['Survived']

x_test = test_data
y_test = pd.DataFrame()

```

```python
x_train["Age"].describe()
```

```python
features = ['Pclass', 'Sex', 'SibSp', 'Parch']

x_train = pd.get_dummies(train_data[features])
x_test = pd.get_dummies(test_data[features])
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

modelRF = RandomForestClassifier(n_estimators = 100, max_depth = 5)

modelRF.fit(x_train,y_train)
```

```python
predictions = modelRF.predict(x_test)
predictions
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
#0.77272 score
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

modelGBC = GradientBoostingClassifier()
modelGBC.fit(x_train, y_train)
```

# CatBoost

```python
from catboost import CatBoostClassifier

x_train_cat = train_data[features + ["Age"]]
y_train_cat = train_data["Survived"]

x_test_cat = test_data[features + ["Age"]]

modelCatBoost = CatBoostClassifier(cat_features = features)

modelCatBoost.fit(x_train_cat,y_train_cat)
```

```python
y_test_cat = modelCatBoost.predict(x_test_cat)
y_test_cat
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_test_cat})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
